### Oxygen uptake analysis


We decided to take a closer look at the changes in yeast metabolism, based on our model, that occur under the influence of a change in the intensity of oxygen uptake. We decided to compare the changes in possible flux ranges under low oxygen conditions (where the relationship examined in analysis 5 was linear) and under high oxygen conditions, which led to a decrease in product flux concentration. For this purpose, we used the flux variability analysis. For this purpose, we used flux variability analysis. Displayed below for presentation are the first 300 reactions, the ranges of which were compared under normal (normal model) and high (model 4, in which the oxygen uptake concentration was maximised) conditions.

```python
from cobra.io import read_sbml_model
import pandas as pd
pd.set_option('display.max_rows', None)
model = read_sbml_model('/Users/Admin/github-classroom/27410/group-assingment-team-4/yeast-GEM-8.7.0/model/yeast-GEM.xml')
from cobra.flux_analysis import flux_variability_analysis
flux_variability_analysis(model, model.reactions[:300])
```

```python
model2 = read_sbml_model('/Users/Admin/github-classroom/27410/group-assingment-team-4/yeast-GEM-8.7.0/model/yeast-GEM4.xml')
from cobra.flux_analysis import flux_variability_analysis
flux_variability_analysis(model2, model2.reactions[:300])
```

Results shows, that some reactions fluxes are not present when oxygen uptake is high, when other become obligatory. Most notable changes are that alanyl-tRNA synthetase reaction is not longer present (r_0157) and deoxycytidine deaminase reaction is now obligatory (r_0329). Deoxycytidine deaminase is involved in salvage of both exogenous and endogenous cytidine and 2'-deoxycytidine for UMP (uridine monophosphate).
